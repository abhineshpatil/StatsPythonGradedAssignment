```python
#Task 3 : Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)
scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 0]['MEDV'],
                   boston_df[boston_df['CHAS'] == 1]['MEDV'], equal_var = True)
```

```python
# Conclusion: Since p-value is less than alpha value of 0.05, so we reject null hypothesis as there is enough proof
# that there is a statistical difference in median value of houses bounded by the Charles river or not
```

```python
#Task 3 : Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)
boston_df.loc[(boston_df['AGE'] <= 35), 'age_group'] = 'age 35 years and younger'
boston_df.loc[(boston_df['AGE'] > 35)&(boston_df['AGE'] < 70), 'age_group'] = 'age between 35 and 70 years'
boston_df.loc[(boston_df['AGE'] >= 70), 'age_group'] = 'age 70 years and older'

a=boston_df[boston_df['age_group'] == 'age 35 years and younger']['MEDV']
b=boston_df[boston_df['age_group'] == 'age between 35 and 70 years']['MEDV']
c=boston_df[boston_df['age_group'] == 'age 70 years and older']['MEDV']

f_statistic, p_value = scipy.stats.f_oneway(a, b, c)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))
```

```python
#Conclusion: Since p value is less than alpha value of 0.05, so we reject null hypothesis as there is significant evidence that at least one of the means differ.
```
